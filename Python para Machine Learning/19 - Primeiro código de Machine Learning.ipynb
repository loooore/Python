```python
import pandas as pd

vinhos_arquivo = pd.read_csv("C:/Users/l50038499/Desktop/wine_dataset.csv")
```

```python
vinhos_arquivo.head(10000)
```

```python
vinhos_arquivo['style'] = vinhos_arquivo['style'].replace('red', 0)
```

```python
vinhos_arquivo['style'] = vinhos_arquivo['style'].replace('white', 1)
```

```python
# Separando as variÃ¡veis entre preditoras e variÃ¡veis alvo
alvo = vinhos_arquivo['style']
resto = vinhos_arquivo.drop('style', axis = 1) # excluindo a coluna style / axis = 1 - significa q Ã© operaÃ§Ãµes na coluna
```

```python
from sklearn.model_selection import train_test_split

# Criando os conjuntos de dados de treino e teste:
resto_treino, resto_teste, alvo_treino, alvo_teste = train_test_split(resto, alvo, test_size = 0.3) # 30% dos dados vÃ£o ser pra teste, 
                                                                                                    # os 70% vÃ£o ser pra treino
```

```python
from sklearn.ensemble import ExtraTreesClassifier

# CriaÃ§Ã£o de modelo:
modelo = ExtraTreesClassifier()
modelo.fit(resto_treino, alvo_treino)

# Imprimindo resultados:
resultado = modelo.score(resto_teste, alvo_teste)
print('AcurÃ¡cia:', resultado)
```
